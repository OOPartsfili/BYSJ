```python
import os
import pandas as pd
import matplotlib.pyplot as plt

def plot_scores_from_folder(folder_path, output_folder='plots'):
    # 获取原文件夹倒数第二级名称
    folder_parts = os.path.normpath(folder_path).split(os.sep)
    folder_base = folder_parts[-2] if len(folder_parts) >= 2 else folder_parts[-1]
    # 将输出文件夹命名为 'plots_倒数第二级原文件夹名'
    output_folder_unique = f"{output_folder}_{folder_base}"
    # 如果输出文件夹不存在，则创建它
    if not os.path.exists(output_folder_unique):
        os.makedirs(output_folder_unique)

    # 遍历文件夹下的所有文件
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.csv'):
            file_path = os.path.join(folder_path, file_name)
            
            try:
                # 读取CSV文件
                df = pd.read_csv(file_path)

                # 检查必要的列是否存在
                if 'sim_time' not in df.columns or 'score' not in df.columns:
                    print(f"跳过文件 {file_name}: 缺少 'sim_time' 或 'score' 列")
                    continue

                # 创建画布
                plt.figure(figsize=(10, 6))
                
                # 绘制折线图
                plt.plot(df['sim_time'], df['score'], marker='o', linestyle='-', label='Score', color='b', markersize=4)

                # 计算横轴范围以绘制三等分的垂直虚线
                min_time = df['sim_time'].min()
                max_time = df['sim_time'].max()
                time_range = max_time - min_time

                if time_range > 0:
                    v1 = min_time + time_range / 3
                    v2 = min_time + 2 * time_range / 3
                    
                    # 绘制两条垂直虚线
                    plt.axvline(x=v1, color='red', linestyle='--', alpha=0.7, label='1/3 Segment')
                    plt.axvline(x=v2, color='red', linestyle='--', alpha=0.7, label='2/3 Segment')

                # 设置图表标题和标签
                plt.title(f'Score Trend: {file_name}')
                plt.xlabel('Simulation Time (sim_time)')
                plt.ylabel('Score')
                plt.legend()
                plt.grid(True, alpha=0.3)

                # 保存图片，文件名与原CSV文件对应
                output_path = os.path.join(output_folder_unique, f"{os.path.splitext(file_name)[0]}.png")
                plt.savefig(output_path)
                plt.close() # 释放内存
                
                print(f"成功绘制并保存: {output_path}")

            except Exception as e:
                print(f"处理文件 {file_name} 时出错: {e}")

# 使用示例
# 假设你的文件夹名为 'score_data'
plot_scores_from_folder(r'data\第二次预实验\cl\score_data')
```
